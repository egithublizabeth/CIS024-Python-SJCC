```python
# Collecting and Parsing a Web Page

import requests
from bs4 import BeautifulSoup

#Collect the webpage contents and put it in variable "table"
webpage = requests.get('https://www.tdcj.state.tx.us/death_row/dr_executed_offenders.html')

#Create a BeautifulSoup object that is parsed(analyze sentence in parts)
soup = BeautifulSoup(webpage.text, 'html.parser')

#Collect the table that I want
# Pull all text from the tdcj_table indent
offender_data = soup.find(class_='tdcj_table indent')

#collect links for offender info and last statement
offender_links = offender_data.find_all('a') 

#Create a for loop to print out all offender's data
for offender_info in offender_data:
        print(offender_info)

```
