```python
from bs4 import BeautifulSoup
import urllib.request
import csv

url = 'https://www.tdcj.state.tx.us/death_row/dr_executed_offenders.html'

html = urllib.request.urlopen(url).read().decode()
soup = BeautifulSoup(html, 'html.parser')
table = soup.select_one("table.tdcj_table")

# python3 just use th.text
headers = [th.text.encode("utf-8") for th in table.select("tr th")]

with open("offender_table.csv", "w") as f:
    wr = csv.writer(f)
    wr.writerow(headers)
    wr.writerows([[td.text.encode("utf-8") for td in row.find_all("td")] for row in table.select("tr + tr")])
    
```

```python
#Count per Race

#Read the contents of the file
file = open("offender_table.csv", "r")

#Initialize an empty dictionary to maintain the counts of each word
raceDict = { }

#Skip the first line
file.readline()

#Creating a new empty list to hold the race values later on
raceList = []

#Creating a for loop to split each element and extract race
for i in file:
    
    # Split each line in the file to get the race
    split = i.split(",") 
    
    #grab the age values
    race = split[-2]
    
    #remove 'b'
    clean = race.replace("b", "")
   
    #Create a new list with all the races
    raceList.append(clean)

 
for word in raceList:
    #If the word is new, then add to dictionary and initialize count to 1 for this word
    if raceDict.get(word) == None:
        raceDict[word] = 1
    else:
    #If the word is already in the dictionary, then increment the count of that word by 1
        raceDict[word] += 1

#Sort dictionary
sortedDict = sorted(raceDict, key=raceDict.get, reverse = True)

# Loop through all the keys in the dictionary and grab the first 10 
for key in sortedDict[0:11]:
    print(key, raceDict[key])

file.close()
### END CODE
```

```python
# Collecting and Parsing a Web Page

import requests
from bs4 import BeautifulSoup

#Collect the webpage contents and put it in variable "table"
webpage = requests.get('https://www.tdcj.state.tx.us/death_row/dr_executed_offenders.html')

#Create a BeautifulSoup object that is parsed(analyze sentence in parts)
soup = BeautifulSoup(webpage.text, 'html.parser')

#Collect the table that I want
# Pull all text from the tdcj_table indent
offender_data = soup.find(class_='tdcj_table indent')

#collect links for offender info and last statement
offender_links = offender_data.find_all('a') 

#Create a for loop to print out all offender's data
for offender_info in offender_data:
        print(offender_info)
```

```python
#collect links for offender info and last statement
offender_links = offender_data.find_all('a') 

links_list = []

#Grab the links for Executed information and last statement
for link in offender_links[0:]:
        links = 'https://www.tdcj.state.tx.us/death_row/' + link.get('href')
        links_list.append(links)
    
print(links_list)
```

```python
#extract only the last statement links, every even number in the list
#https://stackoverflow.com/questions/12433695/extract-elements-of-list-at-odd-positions

Last_list = links_list[1::2]

print(Last_list)
```

```python
#Create a loop to append all last statements into 1 csv file

#this code works but takes longer than 2 mins to gather info ~ 5- 7 mins to gather data
# I did not want to overide the original csv so check out the csv from the folder for output

from bs4 import BeautifulSoup
from bs4 import SoupStrainer
import urllib.request
import csv

#Last_list -> list of last statments
broken_list = ['https://www.tdcj.state.tx.us/death_row/dr_info/cardenasrubenlast.html', 'https://www.tdcj.state.tx.us/death_row/dr_info/pruettrobertlast.html']

new_file = open("Last_Statements.csv", "w")

for i in Last_list[3:]: #the first two links are broken

    #url = i

    only_p_tags = SoupStrainer("p")

    html_statement = urllib.request.urlopen(i).read().decode()
    soup = BeautifulSoup(html_statement, 'html.parser', parse_only=only_p_tags)

    with open("Last_Statements.csv", "a") as f:
        wr = csv.writer(f)
        wr.writerows(soup)
    
    #f.close()

for i in broken_list: 

    #url = i

    only_p_tags = SoupStrainer("p")

    html_statement = urllib.request.urlopen(i).read().decode()
    soup = BeautifulSoup(html_statement, 'html.parser', parse_only=only_p_tags)

    with open("Last_Statements.csv", "a") as f:
        wr = csv.writer(f)
        wr.writerows(soup)
    
    f.close()

file = open("Last_Statements.csv", "r")
    
print(file.read())

file.close()
```

```python
import csv
from textblob import TextBlob

infile = '/Users/elizabeth/Desktop/Python SJCC Fall17/Final/Last_Statements.csv'

polarity_list = []
subjectivity_list = []

with open(infile, 'r') as csvfile:
    rows = csv.reader(csvfile)
    for row in rows:
        sentence = row[0]
        blob = TextBlob(sentence)
        polarity_list.append(blob.sentiment.polarity)
        subjectivity_list.append(blob.sentiment.subjectivity)
        #print sentence
        #print (blob.sentiment.polarity),blob.sentiment.subjectivity)
        #print(polarity_list)
        #print(subjectivity_list)
        
sum_p = (sum(polarity_list))
sum_s = (sum(subjectivity_list))
len_p = (len(polarity_list))
len_s = (len(subjectivity_list))

print("Polarity range [-1, 1], -1 is negative, 0 is neutral, and 1 is positive.")
print("Together, All the Last Statements Polarity is: %s " %(sum_p/len_p))
print()
print("Subjectivity range [0.0, 1.0], where 0.0 is very objective and 1.0 is very subjective.")
print("Together, All the Last Statements Subjectivity is: %s" %(sum_s/len_s))

```
